### Ejercicio:

Si $U \sim unif(0,1)$

$$F_{X}^{-1}(U)=\frac{1}{\lambda }(-Ln(1-U))^{\frac{1}{\alpha }} \sim Weibull$$

Para demostrar que si $U \sim U(0,1)$, entonces $\frac{1}{\lambda} (-\ln(1-
U))^{\frac{1}{\alpha}} \sim \text{Weibull}(\alpha, \lambda)$, consideramos lo siguiente:

\begin{align*}
F_Y(y) &= P\left( \frac{1}{\lambda} (-\ln(1-U))^{\frac{1}{\alpha}} \leq y \right) \\
&= P\left( (-\ln(1-U))^{\frac{1}{\alpha}} \leq \lambda y \right) \\
&= P\left( -\ln(1-U) \leq (\lambda y)^\alpha \right) \\
&= P\left( \ln(1-U) \geq -(\lambda y)^\alpha \right) \\
&= P\left( 1-U \geq e^{-(\lambda y)^\alpha} \right) \\
&= P\left( U \leq 1 - e^{-(\lambda y)^\alpha} \right) \\
&= 1 - e^{-(\lambda y)^\alpha}.
\end{align*}

concluimos que $Y \sim \text{Weibull}(\alpha, \lambda)\blacksquare $.

# Variables aleatorias (abolutamente) continuas

A la probabilidad $\text{cdf}(x):=P(X\leq x)$ se llama la distribución acumulativa (hasta el valor $x$) de $f(x)$.

Con la notación anterior,
$$P(a\leq X\leq b)=\text{cdf}(b)-\text{cdf}(a)$$
y
$$\text{sf}(a):=P(X>a)=1-\text{cdf}(a).$$

### Cálculo de probabilidades:
$\mathbb{P}(0<Z<b)$: Queremos calcular $\mathbb{P}(0<Z<0.43)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha: $\mathbb{P}(0<Z<0.43) = 0.1664$



```python
norm(0,1).cdf(0.43)-norm(0,1).cdf(0)
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-3,3,0.001)
ax.plot(x, norm.pdf(x,loc=0,scale=1))
ax.set_title("N(0,$1^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
px=np.arange(0,0.43,0.01)
ax.set_ylim(0,0.45)
ax.fill_between(px,norm.pdf(px,loc=0,scale=1),alpha=0.5, color='g')
pro=norm(0,1).cdf(0.43)-norm(0,1).cdf(0)
ax.text(0.2,0.02,round(pro,4), fontsize=20)
plt.show()
```

$\mathbb{P}(-b<Z<b)$: Queremos calcular $\mathbb{P}(-0.16<Z<0.16)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha:  $\mathbb{P}(-0.16<Z<0.16) = \mathbb{P}(-0.16<Z<0)+\mathbb{P} (0<Z<0.16) = \mathbb{P}(0<Z<0.16)+\mathbb{P}(0<Z<0.16) = 2\mathbb{P}(0<Z<0.16) = 2(0.0636) = 0.1272$



```python
from scipy.stats import norm

# Valores Z
z = 0.16

# Calcula la probabilidad P(-0.16 < Z < 0.16)
prob = 2 * (norm.cdf(z) - norm.cdf(0))
print(prob)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Configuración de la gráfica
fig, ax = plt.subplots()

# Curva de la distribución
x = np.arange(-3, 3, 0.001)
ax.plot(x, norm.pdf(x, loc=0, scale=1))
ax.set_title("N(0,$1^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

# Área sombreada para la probabilidad \mathbb{P}(-0.16 < Z < 0.16)
px = np.arange(-0.16, 0.16, 0.001)
ax.set_ylim(0, 0.45)
ax.fill_between(px, norm.pdf(px, loc=0, scale=1), alpha=0.5, color='g')

# Cálculo de la probabilidad
pro = 2 * (norm.cdf(0.16) - norm.cdf(0))
ax.text(0, 0.02, round(pro, 4), fontsize=20, ha='center')

plt.show()
```

$\mathbb{P}(Z<-b)$: Queremos calcular $\mathbb{P}(Z<-1.94)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha: $\mathbb{P}(Z<-1.94) = \mathbb{P}(Z<0)+\mathbb{P}(-1.94<Z<0) = \mathbb{P}(Z<0) + \mathbb{P}(0<Z<1.94) = 0.5 - 0.4738 = 0.0262$



```python
pro = norm.cdf(-1.94)
print(f'P(Z < -1.94) = {pro:.4f}')
```

```python
fig, ax = plt.subplots()

# Curva de la distribución
x = np.arange(-3, 3, 0.001)
ax.plot(x, norm.pdf(x, loc=0, scale=1))
ax.set_title("N(0,$1^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

# Área sombreada para la probabilidad \mathbb{P}(Z < -1.94)
px = np.arange(-3, -1.94, 0.001)
ax.set_ylim(0, 0.45)
ax.fill_between(px, norm.pdf(px, loc=0, scale=1), alpha=0.5, color='g')

# Cálculo de la probabilidad
pro = norm.cdf(-1.94)
ax.text(-2.5, 0.05, f'P(Z < -1.94) = {round(pro, 4)}', fontsize=12)

plt.show()
```

$\mathbb{P}(Z>-b)$: Queremos calcular $(Z>-0.07)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha: $\mathbb{P}(Z>-0.07) = \mathbb{P}(-0.07<Z<0) + \mathbb{P}(Z>0) = \mathbb{P}(0<Z<0.07) + \mathbb{P}(Z>0) = 0.0279 + 0.5 = 0.5279$

```python
prob = 1 - norm.cdf(-0.07)
print(f'P(Z > -0.07) = {prob:.4f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

fig, ax = plt.subplots()

# Curva de la distribución
x = np.arange(-3, 3, 0.001)
ax.plot(x, norm.pdf(x, loc=0, scale=1))
ax.set_title("N(0,$1^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

# Área sombreada para la probabilidad \(\mathbb{P}(Z > -0.07)\)
px = np.arange(-3, -0.07, 0.001)
ax.set_ylim(0, 0.45)
ax.fill_between(px, norm.pdf(px, loc=0, scale=1), alpha=0.5, color='g')

# Cálculo de la probabilidad
pro = 1 - norm.cdf(-0.07)
ax.text(-2, 0.1, f'P(Z > -0.07) = {round(pro, 4)}', fontsize=12, ha='center')

plt.show()
```
