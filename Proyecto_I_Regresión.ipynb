Encontrar $\beta_0$ y $\beta_1$ en una regresion lineal

Dada una regresion lineal simple

$y=\beta_0 +\beta_1 x+ \epsilon$,

donde

- $y$ es la variable dependiente,
- $x$ es la variable independiente,
- $\beta_0$ es el intercepto,
- $\beta_1$ es el coeficiente de pendiente,
- $\epsilon$ es el termino error.

Vamos a definir las variables para la Casta Rica

```python
# @title
X = pd.DataFrame({'beta0' : np.ones(CR.shape[0]), 'beta1' : CR['W calculado']})
X
```

```python
# @title
y= CR['Longitud']
y
```

Ahora vamos a implementar una regresion lineal usando el metodo de Minimos Cuadrados Ordinarios (OLS)

```python
# @title
#opcion para sustituir 'x'
#ols: funcion para inimos cuadrados ordinarios
#x_c intercepto = sm.add_constant(x)
#especificamos el modelo a usar
modelo=sm.OLS(y,X)

#Ajustamos el modelo con fit
resultado=modelo.fit()

#imprimimos los resultados
print(resultado.summary())
```

# Ejercicio 2

En ocasiones es adecuado aplicar transformaciones a los datos, investigue y comente: cuántas, cuáles y la utilidad de cada una de ellas.

Las transformaciones en análisis de regresión son técnicas que se aplican a los datos con el fin de mejorar el ajuste del modelo, manejar relaciones no lineales, corregir problemas de heterocedasticidad (varianza no constante), o cuando los datos no cumplen con los supuestos de normalidad. A continuación, se describen algunos tipos comunes de transformaciones y sus usos.

### 1. Transformaciones Logarítmicas

Propósito:

Las transformaciones logarítmicas son útiles para manejar datos con distribuciones sesgadas y para transformar relaciones exponenciales en relaciones lineales, facilitando su análisis.

Ejemplo:

Si la relación entre las variables es de tipo exponencial, es decir,
$$
Y = \beta_0 \cdot X^{\beta_1},
$$
tomar el logaritmo de ambos lados convierte la relación en una forma lineal:
$$
\log(Y) = \log(\beta_0) + \beta_1 \log(X).
$$
Esto permite aplicar técnicas de regresión lineal a relaciones originalmente no lineales.

Aplicaciones:

- Corregir asimetrías en la distribución de los datos.
- Reducir la variabilidad de los datos y controlar la heterocedasticidad.
- Facilitar la interpretación de los coeficientes en términos de elasticidades (en economía).

### 2. Transformaciones de Potencia

Propósito:

Las transformaciones de potencia, como la raíz cuadrada o el inverso, son útiles para estabilizar la varianza y tratar relaciones no lineales entre variables.

Ejemplos:

- **Raíz cuadrada:** Si los datos muestran una relación cuadrática o una distribución positiva sesgada, la raíz cuadrada puede ser aplicada a las variables:
  $$
  Y' = \sqrt{Y}
  $$
  Esto tiende a reducir la asimetría y hacer que la varianza sea más constante.

- **Transformación inversa:** Si la relación entre las variables es hiperbólica, se puede aplicar la transformación inversa:
  $$
  Y' = \frac{1}{Y} \quad \text{y} \quad X' = \frac{1}{X}.
  $$
  Esto puede convertir una relación no lineal en una más lineal.

Aplicaciones:

- Reducir la heterocedasticidad.
- Linealizar relaciones no lineales.
- Tratar problemas de distribución sesgada.

### 3. Transformaciones Polinómicas

Propósito:

Las transformaciones polinómicas se utilizan para modelar relaciones no lineales entre las variables dependientes e independientes mediante la adición de términos polinómicos.

Ejemplo:

Un modelo de regresión polinómica de segundo grado puede capturar una relación cuadrática entre \( X \) y \( Y \):
$$
Y = \beta_0 + \beta_1 X + \beta_2 X^2.
$$
Este tipo de modelo es útil cuando se sospecha que la relación entre las variables tiene una forma curva.

Aplicaciones:

- Modelar relaciones no lineales (curvatura).
- Mejorar el ajuste del modelo cuando una línea recta no es adecuada.

### 4. Transformación Logit

Propósito:

La transformación logit se utiliza principalmente en regresión logística para modelar variables dependientes binarias, donde el resultado es 0 o 1.

Ejemplo:

El modelo logit transforma la probabilidad de éxito en una variable continua que puede ser modelada linealmente:
$$
\text{logit}(p) = \log\left(\frac{p}{1-p}\right).
$$
Donde \( p \) es la probabilidad de que el evento ocurra.

Aplicaciones:

- Modelar resultados binarios (sí/no, éxito/fallo).
- Interpretar los coeficientes en términos de razones de probabilidades (odds ratios).

### 5. Transformación Box-Cox

Propósito:

La transformación Box-Cox es una técnica de transformación de potencia que busca estabilizar la varianza y aproximar la distribución de los datos a la normalidad.

Ejemplo:

La transformación Box-Cox se define como:
$$
Y(\lambda) = \frac{Y^\lambda - 1}{\lambda}, \quad \text{si} \, \lambda \neq 0
$$
$$
Y(\lambda) = \log(Y), \quad \text{si} \, \lambda = 0.
$$
El parámetro \(\lambda\) se estima a partir de los datos y determina la naturaleza de la transformación (logarítmica, inversa, etc.).

Aplicaciones:

- Estabilización de la varianza en presencia de heterocedasticidad.
- Transformación de datos para aproximar la normalidad.
- Mejora del ajuste del modelo y de las propiedades inferenciales.

### 6. Transformación de Escala

Propósito:

Las transformaciones de escala, como la estandarización y normalización, son técnicas que ajustan las variables para que se encuentren en un rango específico, lo cual es útil cuando se comparan variables en diferentes escalas o se usan en modelos de aprendizaje automático.

Ejemplos:

- **Estandarización:** Restar la media y dividir por la desviación estándar:
  $$
  X' = \frac{X - \mu_X}{\sigma_X}.
  $$
  Esto centra los datos alrededor de 0 con una desviación estándar de 1.

- **Normalización:** Ajustar los datos para que se encuentren entre 0 y 1:
  $$
  X' = \frac{X - \min(X)}{\max(X) - \min(X)}.
  $$
  Esto facilita la comparación entre variables que originalmente tenían diferentes rangos.

Aplicaciones:

- Mejora de la convergencia en algoritmos de optimización.
- Facilita la interpretación de los coeficientes en modelos con múltiples variables.
- Prevención de sesgos en modelos de aprendizaje automático.

### 7. Transformaciones de Variables Categóricas

Propósito:

Las transformaciones de variables categóricas permiten incluir variables categóricas en un modelo de regresión mediante la creación de variables dummy o indicadores.

Ejemplo:

Si una variable tiene tres categorías (por ejemplo, "Bajo", "Medio", "Alto"), se pueden crear dos variables dummy para representarlas:
- Dummy1: \( 1 \) si "Medio", \( 0 \) en caso contrario.
- Dummy2: \( 1 \) si "Alto", \( 0 \) en caso contrario.

La categoría "Bajo" se convierte en la categoría de referencia.

Aplicaciones:

- Incluir variables categóricas en modelos de regresión.
- Interpretar los efectos de categorías cualitativas sobre la variable dependiente.
- Mejorar la capacidad explicativa del modelo cuando se incluyen variables no numéricas.

### 8. Transformaciones Recíprocas

Propósito:

Las transformaciones recíprocas son útiles para manejar relaciones no lineales en las que el cambio en una variable es inversamente proporcional al cambio en otra.

Ejemplo:

Una transformación recíproca se aplica a una variable para modelar relaciones de tipo inverso:
$$
Y = \frac{1}{X}.
$$
Esto puede ser útil cuando se espera que la relación entre \( Y \) y \( X \) disminuya a medida que \( X \) aumenta.

Aplicaciones:

- Modelar relaciones inversas entre variables.
- Corregir relaciones no lineales para aplicar modelos lineales.

### 9. Transformaciones Exponenciales

Propósito:

Las transformaciones exponenciales se utilizan para modelar relaciones donde el cambio en la variable dependiente es proporcional al valor actual de la misma, lo que es común en procesos de crecimiento exponencial.

Ejemplo:

Un modelo de crecimiento exponencial puede expresarse como:
$$
Y = \beta_0 e^{\beta_1 X}.
$$
Este tipo de transformación es útil en situaciones donde las tasas de crecimiento o decaimiento son constantes en términos relativos.

Aplicaciones:

- Modelar procesos de crecimiento exponencial o decaimiento.
- Capturar la naturaleza multiplicativa de algunas relaciones entre variables.
- Aplicar técnicas de regresión a datos con crecimiento o decrecimiento exponencial.

**- b) El estadístico de prueba se distribuye Normal, ¿Cuál es el nombre de esta prueba y cuál es la fórmula del estadístico de prueba de la misma?**

El estadístico de prueba que se distribuye normal es conocido como el estadístico de Wald, este se utiliza para probar la significancia de los coeficientes del modelo.

Su fórmula es:
\begin{equation}
T = \frac
    { ( \hat{\theta} - \theta_{0} )^2 }
    {Var(\hat{\theta})}
\end{equation}

Donde:

$\hat{\theta} : $ estimadores de los parámetros.

$\theta_{0} : $ son los valores de los parámetros propuestos, el bajo del parámetro bajo la hipótesis nula.

En otras palabras en el númerador se pone la diferencia entre el estimador y el valor bajo la hipótesis nula. Mientras, el denominador es el error estándar del estimador.


**- c) explica: el seudo $R^2$ de McFadden, $R^2$ de Cox-Snell y $R^2$ de Nagelkerke.**
  
  $R^2$ de McFadden: Es una medida utilizada en modelos de regresión logística para evaluar la bondad de ajuste del modelo. A diferencia del $R^2$ en la regresión lineal, que mide la proporción de la varianza explicada por el modelo, el $R^2$ de McFadden se basa en la comparación de la verosimilitud del modelo ajustado con la verosimilitud de un modelo nulo (un modelo sin predictores). La fórmula es:

$$ R^2_{McFadden} = 1 - \frac{Ln(L_{modelo})}{Ln(L_{nulo})} $$

Donde:

$\ln(L_{\text{modelo}}) : $ es el logaritmo de la verosimilitud del modelo ajustado.

$\ln(L_{\text{nulo}}) : $ es el logaritmo de la verosimilitud del modelo nulo.


  $R^2$ de Cox-Snell: Esta medida se basa en la prueba de razón de verosimilitud del modelo nulo y el modelo que está siendo ajustado. Cuanto mejor sea el modelo que está siendo ajustado en comparación con el modelo nulo, menor será la relación entre $L_{nulo}$ y $L_{modelo}$. La fórmula es:

$$ R^2_{CoxSnell} = 1 - \left(\frac{L_{nulo}}{L_{modelo}}\right)^{2/n} $$

Donde:

$n : $ es el tamaño de la muestra.

$R^2$ de Nagelkerke: Esta medida es una extensión de la medida de Cox-Snell, ajusta la escala de la estadística para cubrir todo el rango de 0 a 1, haciéndola más interpretable. El pseudocoeficiente de determinación de Nagelkerkes se convierte en 1 si el modelo que está siendo ajustado da una predicción perfecta con una probabilidad de 1.

 Se calcula como:

$$ R^2_{Nagelkerke} = \frac{R^2_{CoxSnell}}{max(R^2_{CoxSnell})} = \frac{1 - \left(\frac{L_{nulo}}{L_{modelo}}\right)^{2/n}}{1 - (L_{nulo}
)^{2/n}} $$

Dadas dos variables aleatorias $X$ y $Y$ que tienen función de densidad conjunta $f(x, y) = xe^{−x(1+y)}$ para $x > 0$, $y > 0$ . Encontrar la ecuación de regresión de $Y$ sobre $X$ y bosqueje la curva de regresión.

Para poder encontrar la ecuación de regresión primero debemos de encontrar la función de densidad marginal de $X$, esta se obtiene integrando la función de densidad conjunta sobre los valores posibles de $Y$.

\begin{equation}
f_X (x)=∫_0^∞ f(x,y) dy=∫_0^∞ xe^{(-x(1+y))} dy
\end{equation}

\begin{equation}
f_X (x)=xe^{-x} ∫_0^∞ e^{-xy} dy = xe^{-x} \left[\frac{e^{-xy}}{-x}\right]_0^∞ = xe^{-x} \left(0-\left(\frac{-1}{x}\right)\right)=e^{-x}
\end{equation}

Después encontramos la función de densidad condicional de $Y$ dado $X$, esta la obtenemos dividiendo la función de densidad conjunta por la función de densidad marginal de $X$:

\begin{equation}
E(Y|X=x) = ∫_{0}^{∞} yf_{Y|X}(y|x)dy = ∫_{0}^{∞}yxe^{-xy}dy
\end{equation}

Resolvemos la integral por partes:

Sea $u=y$ y $dv=xe^{-xy} dy$

Entonces, $du = dv$ y $v = -e^{-xy}$

\begin{equation}
∫_{0}^{∞}yxe^{-xy}dy = [-ye^{-xy}]_{0}^{∞} + ∫_0^∞ e^{-xy}dy = 0 + ∫_0^∞ e^{-xy}dy = \frac{1}{x}
\end{equation}

Por lo tanto:

\begin{equation}
E(Y|X=x) = \frac{1}{x}
\end{equation}

La ecuación de regresión de $Y$ sobre $X$ es

\begin{equation}
Y = \frac{1}{X}
\end{equation}

```python
# @title
# Importamos las bibliotecas necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos los datos
# np.linspace crea un array de 400 puntos entre -10 y 10
x = np.linspace(-10, 10, 400)

# Calculamos Y = 1/X
# Usamos np.errstate para manejar la división por cero y evitar warnings
with np.errstate(divide='ignore', invalid='ignore'):
    y = 1 / x

# Creamos el gráfico
plt.plot(x, y, label='Y = 1/X')  # Graficamos Y = 1/X y añadimos una etiqueta

# Añadimos etiquetas a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Añadimos un título al gráfico
plt.title('Gráfico de la ecuación de regresión de Y sobre X')

# Limitamos el eje Y para evitar valores extremos debido a la discontinuidad en X = 0
plt.ylim(-10, 10)

# Añadimos una cuadrícula para mejor visualización
plt.grid(True)

# Añadimos líneas en los ejes X e Y para mayor claridad
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Añadimos una leyenda
plt.legend()

# Mostramos el gráfico
plt.show()
```

La ecuación de regresión a ecuación de regresión $E(Y|X=x) = \frac{1}{x}$ nos dice que la esperanza condicional de $Y$ dado $X$ es inversamente proporcional a $X$. Esto significa que a medida que $X$ aumenta, el valor esperado de $Y$ disminuye.

Con base en las características de la curva podemos observar que la forma es de una hiérbola que decrece a medida que $X$ aumenta.

Para valores grandes de $X$, $Y$ es pequeño.

Cuando $X$ es muy grande, $Y$ se aproxima a $0$.

### Comparación entre Modelo Saturado y Modelo con Subconjunto de Variables

### Modelo Saturado

Un modelo saturado incluye todas las posibles variables y términos de interacción en el análisis. En teoría, un modelo saturado puede ajustar perfectamente los datos, ya que tiene suficiente flexibilidad para capturar todas las relaciones presentes en los datos. Sin embargo, este tipo de modelo tiene varias desventajas:

1. **Sobreajuste (Overfitting)**:
   - **Definición**: El sobreajuste ocurre cuando un modelo se ajusta demasiado a los datos de entrenamiento, capturando el ruido en lugar de la señal subyacente. Esto puede resultar en un desempeño deficiente en datos nuevos.
   - **Consecuencia**: Aunque el modelo saturado puede tener un buen desempeño en el conjunto de entrenamiento, es probable que tenga un desempeño deficiente en el conjunto de prueba o en datos futuros debido a su complejidad y sobreajuste.

2. **Complejidad del Modelo**:
   - **Definición**: Un modelo saturado puede ser extremadamente complejo, con muchas variables y términos de interacción.
   - **Consecuencia**: Esta complejidad puede dificultar la interpretación del modelo y hacer que sea menos práctico en aplicaciones reales. Además, modelos complejos pueden ser más costosos en términos de computación.

3. **Multicolinealidad**:
   - **Definición**: La multicolinealidad ocurre cuando las variables independientes están altamente correlacionadas entre sí.
   - **Consecuencia**: En un modelo saturado, la multicolinealidad puede ser un problema, haciendo que los coeficientes estimados sean imprecisos y difíciles de interpretar.

### Modelo con Subconjunto de Variables

Un modelo con un subconjunto de variables seleccionadas puede ofrecer ventajas significativas en comparación con un modelo saturado:

1. **Prevención del Sobreajuste**:
   - **Definición**: Al limitar el número de variables, el modelo es menos propenso a ajustar el ruido en los datos y tiene menos riesgo de sobreajuste.
   - **Beneficio**: Un modelo con un subconjunto de variables puede generalizar mejor a datos nuevos, proporcionando predicciones más robustas.

2. **Simplicidad e Interpretabilidad**:
   - **Definición**: Un modelo más simple es más fácil de interpretar y entender.
   - **Beneficio**: Los modelos con menos variables y términos son más fáciles de comunicar a los stakeholders y permiten una mejor comprensión de las relaciones entre las variables.

3. **Menor Multicolinealidad**:
   - **Definición**: Al seleccionar un subconjunto de variables, se puede reducir el riesgo de multicolinealidad.
   - **Beneficio**: Menos multicolinealidad lleva a estimaciones más estables y confiables de los coeficientes del modelo.

4. **Eficiencia Computacional**:
   - **Definición**: Modelos más simples suelen requerir menos recursos computacionales.
   - **Beneficio**: Menos variables y términos hacen que el modelo sea más eficiente en términos de tiempo de entrenamiento y predicción.

### Métodos para Selección de Variables

Para decidir qué variables incluir en el modelo, se pueden utilizar métodos como:

- **Selección hacia adelante**: Añadir variables al modelo una por una, evaluando el impacto de cada variable en el desempeño del modelo.
- **Selección hacia atrás**: Comenzar con un modelo completo y eliminar variables que no contribuyen significativamente al modelo.
- **Métodos de regularización**: Como LASSO (Least Absolute Shrinkage and Selection Operator) o Ridge Regression, que penalizan la inclusión de demasiadas variables.

### Conclusión

En general, un modelo con un subconjunto de variables es preferible al modelo saturado debido a su capacidad para evitar el sobreajuste, mejorar la interpretabilidad, reducir la multicolinealidad, y ser más eficiente computacionalmente. Sin embargo, la selección del modelo adecuado debe basarse en el contexto específico del problema, los objetivos del análisis, y la calidad de los datos. Un modelo bien seleccionado y validado es crucial para proporcionar insights precisos y útiles.
